**1. Задание**
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline


v1 = [10, 10, 10]
v2 = [0, 0, -10]

def vector_length(v):
    sum_square = 0
    for i in range(len(v)):
        sum_square += v[i]**2
    return math.sqrt(sum_square)

length_v1 = vector_length(v1)
length_v2 = vector_length(v2)

length_v1, length_v2
```

**3. Задание (в программе)**

Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
#Окружность
plt.gca().set_aspect('equal')
x = []
y1 = []
y2 = []
r = 5
for i in np.arange(-r, r, 0.00001):
    x.append(i)
    y1.append(math.sqrt(r**2 - i**2))
    y2.append(-math.sqrt(r**2 - i**2))
plt.plot(x,y1)   
plt.plot(x,y2)
```

```python
#Эллипс
a = 3
b = 2
x = []
y1 = []
y2 = []
for i in range(-300, 301):
    x_ = i/100
    x.append(x_)
    y1.append(math.sqrt((a**2*b**2-x_**2*b**2)/a**2))
    y2.append(-math.sqrt((a**2*b**2-x_**2*b**2)/a**2))
plt.plot(x,y1);
plt.plot(x,y2);
```

```python
#гипербола

a=1
b=2
xl = []
xr = []
y1 = []
y2 = []
for i in range(100):
    x_ = i/100 + a
    xl.append(-x_)
    xr.append(x_)
    y1.append(np.sqrt(abs(-b**2*(1 - x_**2/a**2))))
    y2.append(-np.sqrt(abs(-b**2*(1 - x_**2/a**2))))
plt.plot(xl, y1)
plt.plot(xl, y2)
plt.plot(xr, y1)
plt.plot(xr, y2)
```

**3. Задание (в программе)**
Напишите код, который будет переводить полярные координаты в декартовы.


```python
def polar_to_decart(a, r):
    a = np.pi/180*a
    x = r*np.cos(a)
    y = r*np.sin(a)
    return (x, y)

polar_to_decart(20, 5)
```

**4. Задание (в программе)**
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
from scipy.optimize import fsolve
def equations(p):
    x, y = p
    return y - x**2 + 1, np.exp(x) + x - x*y - 1

x = np.linspace(-2, 5, 300)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.grid(True)
```

```python
x1, y1 = fsolve(equations, (-2, 1))
x1, y1
```

```python
x2, y2 = fsolve(equations, (2, 5))
x2, y2
```

```python
x3, y3 = fsolve(equations, (10, 11))
x3, y3
```
